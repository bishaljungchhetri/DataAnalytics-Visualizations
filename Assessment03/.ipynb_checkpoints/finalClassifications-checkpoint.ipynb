```python
"""
PRT564 Data Analytics and Visualisation — Assessment 3
Group ID: SYDN G2 (Sydney Campus)

Description:
    This program builds a highly robust, reproducible end-to-end data pipeline 
    to classify public transport network conditions into "High Demand Surge States" 
    versus "Standard Baseline Operations" for Transport for NSW (TfNSW).
    
Requirements Addressed:
    - LO1 & LO3: Complex Exploratory Data Analysis and Cross-Validation Strategy.
    - LO4 & LO5: Advanced Feature Engineering (Lags, rolling windows, date normalization) 
                 and isolated scikit-learn transformers.
    - Deployment: Production-ready endpoint emulation to evaluate live payload strings.
    - HD Enhancements: Dynamic Feature Importance Extraction and Multi-Metric Evaluations.
"""

import os
import warnings
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, StratifiedKFold, GridSearchCV
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve

# Suppress minor environment warnings to maintain a clean console log output
warnings.filterwarnings('ignore', category=UserWarning)
warnings.filterwarnings('ignore', category=FutureWarning)

# Configure clean, professional-grade styling parameters for report inclusion
sns.set_theme(style="whitegrid")
plt.rcParams.update({
    'font.size': 11,
    'axes.labelsize': 12,
    'axes.titlesize': 14,
    'xtick.labelsize': 10,
    'ytick.labelsize': 10,
    'figure.titlesize': 16
})

class TfNSWTransitClassificationPipeline:
    def __init__(self, raw_data_path, output_dir='.'):
        """
        Initializes the pipeline directory paths and system storage structures.
        """
        self.raw_data_path = raw_data_path
        self.output_dir = output_dir
        self.df = None
        self.preprocessor = None
        self.modal_cols = []
        self.grid_models = {}
        self.champion_pipeline = None
        
        # Create output charts and data folders safely in the local workspace directory
        os.makedirs(os.path.join(self.output_dir, 'charts'), exist_ok=True)
        os.makedirs(os.path.join(self.output_dir, 'data'), exist_ok=True)

    def extract_and_clean_data(self):
        """
        Step 1: Ingestion, Dynamic Schema Mapping & Wide-Pivoting (LO4, LO5)
        Pivots rows based on ('Year_Month', 'Travel_Mode', 'Trip') long-format layout.
        """
        print("="*70)
        print("[STAGE 1] INGESTING AND CLEANING SOURCE ARTIFACTS")
        print("="*70)
        
        if not os.path.exists(self.raw_data_path):
            raise FileNotFoundError(f"Missing required input file path: {self.raw_data_path}")
            
        raw_df = pd.read_csv(self.raw_data_path)
        print(f"[*] Raw shape read from disk: {raw_df.shape}")
        print(f"[*] Identified columns: {list(raw_df.columns)}")
        
        # Standardize matching keys to account for variations in long formats
        raw_df.columns = [c.strip() for c in raw_df.columns]
        
        # Map your dataset columns dynamically to ensure predictable parsing
        time_col = 'Year_Month' if 'Year_Month' in raw_df.columns else raw_df.columns[0]
        mode_col = 'Travel_Mode' if 'Travel_Mode' in raw_df.columns else raw_df.columns[2]
        value_col = 'Trip' if 'Trip' in raw_df.columns else raw_df.columns[3]
        
        print(f"[+] Mapping columns -> Index: '{time_col}', Columns: '{mode_col}', Values: '{value_col}'")
        
        # Clean string structures and format types
        raw_df[time_col] = raw_df[time_col].astype(str).str.strip()
        raw_df[mode_col] = raw_df[mode_col].astype(str).str.strip()
        raw_df[value_col] = pd.to_numeric(raw_df[value_col], errors='coerce').fillna(0)
        
        # Pivot from long layout to wide categorical matrix structure
        pivot_df = raw_df.pivot_table(index=time_col, columns=mode_col, values=value_col, aggfunc='sum').reset_index()
        pivot_df.rename(columns={time_col: 'Date'}, inplace=True)
        
        self.df = pivot_df
        print(f"[+] Pivot complete. Shape: {self.df.shape}")
        
        # Standardize dates into unified Datetime coordinates safely without inference warnings
        self.df['Date'] = pd.to_datetime(self.df['Date'], format='mixed', errors='coerce')
        
        # Drop rows where date translation failed completely
        self.df.dropna(subset=['Date'], inplace=True)
        
        # Identify newly created columns matching public transit modes
        all_cols = list(self.df.columns)
        self.modal_cols = [m for m in ['Bus', 'Ferry', 'Light Rail', 'Train', 'Metro'] if m in all_cols]
        
        # Fallback safeguard: Use all numeric columns if standard mode labels aren't found
        if not self.modal_cols:
            self.modal_cols = list(self.df.select_dtypes(include=[np.number]).columns)
            if 'Total' in self.modal_cols: self.modal_cols.remove('Total')
            
        print(f"[*] Aggregating total monthly network trips across components: {self.modal_cols}")
        for col in self.modal_cols:
            self.df[col] = self.df[col].fillna(self.df[col].median())
            
        # Calculate totals correctly to avoid any missing values or metrics issues
        self.df['Total'] = self.df[self.modal_cols].sum(axis=1)
        
        # Sort chronologically to preserve order for lag step calculations
        self.df = self.df.sort_values('Date').reset_index(drop=True)
        print(f"[*] Processed Clean Data Sample Size: {len(self.df)} months remaining.")
        return self.df

    def engineer_advanced_features(self):
        """
        Step 2: Sophisticated Feature Engineering & Robust Windowing (LO4, LO5)
        Applies padding values during shifting to ensure zero rows are accidentally dropped.
        """
        print("\n" + "="*70)
        print("[STAGE 2] ENRICHING METRICS AND BUILDING FEATURE MATRICES")
        print("="*70)
        
        # 1. Temporal feature extractions
        self.df['Month_Num'] = self.df['Date'].dt.month
        self.df['Quarter'] = 'Q' + self.df['Date'].dt.quarter.astype(str)
        self.df['Year'] = self.df['Date'].dt.year
        
        # 2. Time-Lag Feature Processing (Using backfill padding to avoid empty values)
        self.df['Total_Lag_1M'] = self.df['Total'].shift(1).bfill()
        self.df['Total_Lag_2M'] = self.df['Total'].shift(2).bfill()
        
        # 3. Structural Sequence Aggregations (Using min_periods to protect short historical windows)
        self.df['Rolling_Mean_3M'] = self.df['Total'].shift(1).rolling(window=3, min_periods=1).mean().bfill()
        
        # Verify that zero data rows are lost during extraction loops
        print(f"[*] Dataset rows preserved following feature generation: {len(self.df)}")
        
        # 4. Define Classification Target Logic (LO1, LO3)
        # Establish the 65th percentile threshold boundary for systemic "High Demand Surge States"
        surge_threshold = self.df['Total'].quantile(0.65)
        self.df['Demand_Surge_Class'] = (self.df['Total'] >= surge_threshold).astype(int)
        
        print(f"[*] Calculated operational surge state limit: {surge_threshold:,.2f} monthly trips.")
        print(f"[*] Output class balance details: \n{self.df['Demand_Surge_Class'].value_counts(normalize=True)}")
        
        # Save enriched analytical set to output directory
        self.df.to_csv(os.path.join(self.output_dir, 'data', 'enriched_classification_data.csv'), index=False)
        return self.df

    def execute_exploratory_visuals(self):
        """
        Step 3: Advanced Exploratory Data Analysis & Visualizations (LO1, LO2, LO3)
        """
        print("\n" + "="*70)
        print("[STAGE 3] GENERATING PROFESSIONAL VISUALISATIONS AND CORRELATIONS")
        print("="*70)
        
        charts_path = os.path.join(self.output_dir, 'charts')
        
        # Visual 1: Target Class Balance Count plot with explicit hue to satisfy clean output metrics
        plt.figure(figsize=(7, 5))
        ax = sns.countplot(x='Demand_Surge_Class', data=self.df, hue='Demand_Surge_Class', palette='Set2', legend=False)
        plt.title('Target Distribution: Standard Operations vs Demand Surge')
        plt.xlabel('Operational State Category')
        plt.ylabel('Observed Months Count')
        
        ax.set_xticks([0, 1])
        ax.set_xticklabels(['Standard (Class 0)', 'High Surge (Class 1)'])
        for p in ax.patches:
            if p.get_height() > 0:
                ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height() + 0.1), ha='center')
        plt.tight_layout()
        plt.savefig(os.path.join(charts_path, 'v1_target_class_balance.png'), dpi=300)
        plt.close()
        
        # Visual 2: Matrix Correlation Heatmap
        plt.figure(figsize=(10, 8))
        eval_metrics = self.modal_cols + ['Total_Lag_1M', 'Total_Lag_2M', 'Rolling_Mean_3M', 'Demand_Surge_Class']
        corr_cols = [c for c in eval_metrics if c in self.df.columns]
        corr_matrix = self.df[corr_cols].corr()
        sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".3f", linewidths=.5)
        plt.title('Feature Interaction Correlation Heatmap', fontsize=14, pad=15)
        plt.tight_layout()
        plt.savefig(os.path.join(charts_path, 'v2_feature_correlation_matrix.png'), dpi=300)
        plt.close()
        
        # Visual 3: Operational Modality Shifts Boxplot
        plot_modes = [m for m in ['Train', 'Bus'] if m in self.df.columns]
        if not plot_modes and len(self.modal_cols) >= 2:
            plot_modes = self.modal_cols[:2]
            
        if len(plot_modes) >= 2:
            fig, axes = plt.subplots(1, 2, figsize=(14, 6))
            sns.boxplot(x='Demand_Surge_Class', y=plot_modes[0], data=self.df, ax=axes[0], hue='Demand_Surge_Class', palette='Pastel1', legend=False)
            axes[0].set_title(f'{plot_modes[0]} Volume Shifts Across States')
            axes[0].set_xticks([0, 1])
            axes[0].set_xticklabels(['Standard', 'Surge State'])
            axes[0].set_ylabel('Passenger Volume Count')
            
            sns.boxplot(x='Demand_Surge_Class', y=plot_modes[1], data=self.df, ax=axes[1], hue='Demand_Surge_Class', palette='Pastel2', legend=False)
            axes[1].set_title(f'{plot_modes[1]} Volume Shifts Across States')
            axes[1].set_xticks([0, 1])
            axes[1].set_xticklabels(['Standard', 'Surge State'])
            axes[1].set_ylabel('Passenger Volume Count')
            
            plt.suptitle('Primary Transport Mode Volume Distribution Split by Target Class')
            plt.tight_layout()
            plt.savefig(os.path.join(charts_path, 'v3_modal_distribution_boxplot.png'), dpi=300)
            plt.close()
        
        print(f"[*] Visual assets exported successfully to folder: '{charts_path}'")

    def run_isolated_modeling(self):
        """
        Step 4: Preprocessing Pipeline & Stratified Hyperparameter Tuning (LO1, LO4, LO5)
        """
        print("\n" + "="*70)
        print("[STAGE 4] PREPROCESSING PIPELINE INFRASTRUCTURE & ENGINES")
        print("="*70)
        
        num_features = self.modal_cols + ['Total_Lag_1M', 'Total_Lag_2M', 'Rolling_Mean_3M']
        cat_features = ['Quarter']
        
        X = self.df[num_features + cat_features]
        y = self.df['Demand_Surge_Class']
        
        # 75/25 Split stratified by your target to maintain class proportions
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42, stratify=y)
        print(f"[+] Training sample footprint: {X_train.shape[0]} months | Evaluation test footprint: {X_test.shape[0]} months")
        
        # Pipeline transformations
        num_transformer = Pipeline(steps=[('scaler', StandardScaler())])
        cat_transformer = Pipeline(steps=[('encoder', OneHotEncoder(drop='first', handle_unknown='ignore'))])
        
        self.preprocessor = ColumnTransformer(transformers=[
            ('num', num_transformer, num_features),
            ('cat', cat_transformer, cat_features)
        ])
        
        # Cross Validation strategy configuration
        cv_strategy = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)
        
        # --- MODEL 1: LOGISTIC REGRESSION ---
        lr_pipeline = Pipeline(steps=[('prep', self.preprocessor), 
                                      ('clf', LogisticRegression(max_iter=5000, solver='liblinear'))])
        lr_params = {'clf__C': [0.1, 1.0, 10.0], 'clf__penalty': ['l2']}
        print("[*] Tuning regularized Logistic Regression baseline parameters...")
        grid_lr = GridSearchCV(lr_pipeline, lr_params, cv=cv_strategy, scoring='f1', n_jobs=-1).fit(X_train, y_train)
        self.grid_models['Logistic_Regression'] = grid_lr
        
        # --- MODEL 2: RANDOM FOREST ENSEMBLE ---
        rf_pipeline = Pipeline(steps=[('prep', self.preprocessor), 
                                      ('clf', RandomForestClassifier(random_state=42))])
        rf_params = {'clf__n_estimators': [50, 100], 'clf__max_depth': [4, 8]}
        print("[*] Tuning Random Forest Ensemble parameters...")
        grid_rf = GridSearchCV(rf_pipeline, rf_params, cv=cv_strategy, scoring='f1', n_jobs=-1).fit(X_train, y_train)
        self.grid_models['Random_Forest'] = grid_rf
        
        self.champion_pipeline = grid_rf.best_estimator_
        
        # Compute performance output reports
        self._evaluate_performance_metrics(X_test, y_test)
        
        # HD MANDATORY EXTRACTION: Generate Feature Importance Matrix Maps
        self._export_feature_importance_rankings(num_features)

    def _evaluate_performance_metrics(self, X_test, y_test):
        """
        Step 4.2: Evaluates model arrays and handles complex diagnostic plots.
        """
        plt.figure(figsize=(9, 6))
        charts_path = os.path.join(self.output_dir, 'charts')
        
        for name, search_instance in self.grid_models.items():
            best_model = search_instance.best_estimator_
            y_pred = best_model.predict(X_test)
            y_probs = best_model.predict_proba(X_test)[:, 1]
            
            print(f"\n==================== {name.upper()} PERFORMANCE RESULTS ====================")
            print(f"Optimal Hyperparameters: {search_instance.best_params_}")
            print(f"Validation F1-Score: {search_instance.best_score_:.4f}")
            print("\nConfusion Matrix:")
            cm = confusion_matrix(y_test, y_pred)
            print(cm)
            print("\nClassification Matrix Metrics:")
            print(classification_report(y_test, y_pred))
            
            # Export Visual Confusion Matrix Chart to solve structural masking requirements
            plt.figure(figsize=(5, 4))
            sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False,
                        xticklabels=['Standard', 'Surge'], yticklabels=['Standard', 'Surge'])
            plt.title(f"{name.replace('_', ' ')} Confusion Matrix")
            plt.ylabel('Actual Label')
            plt.xlabel('Predicted Label')
            plt.tight_layout()
            plt.savefig(os.path.join(charts_path, f'v4_confusion_matrix_{name.lower()}.png'), dpi=300)
            plt.close()
            
            # Collect data strings for unified ROC curves plot
            plt.figure(1) # Reactivate ROC plot canvas
            fpr, tpr, _ = roc_curve(y_test, y_probs)
            auc_metric = roc_auc_score(y_test, y_probs)
            plt.plot(fpr, tpr, label=f"{name.replace('_', ' ')} (ROC-AUC = {auc_metric:.3f})", lw=2)
            
        plt.figure(1)
        plt.plot([0, 1], [0, 1], 'k--', label='Baseline Guessing (AUC = 0.500)')
        plt.xlabel('False Positive Rate (FPR)')
        plt.ylabel('True Positive Rate (TPR)')
        plt.title('Receiver Operating Characteristic (ROC) Curve Comparison')
        plt.legend(loc='lower right')
        plt.tight_layout()
        plt.savefig(os.path.join(charts_path, 'v4_model_roc_curve.png'), dpi=300)
        plt.close()
        print("\n[*] Diagnostic comparison figures and confusion matrices saved to disk.")

    def _export_feature_importance_rankings(self, num_features):
        """
        Step 4.5: HD Requirement - Extracts Gini Importance Scores from the Ensemble
        """
        print("\n" + "="*70)
        print("[STAGE 4.5] EXTRACTING CHAMPION FEATURE IMPORTANCE WEIGHTS")
        print("="*70)
        
        # Programmatically parse feature columns from the preprocessing pipeline
        cat_encoder = self.champion_pipeline.named_steps['prep'].transformers_[1][1].named_steps['encoder']
        encoded_cat_features = list(cat_encoder.get_feature_names_out(['Quarter']))
        all_feature_names = num_features + encoded_cat_features
        
        # Pull core weight values from the optimized Random Forest classifier
        importances = self.champion_pipeline.named_steps['clf'].feature_importances_
        
        importance_df = pd.DataFrame({
            'Feature': all_feature_names,
            'Importance': importances
        }).sort_values(by='Importance', ascending=False)
        
        print(importance_df.to_string(index=False))
        
        # Plot and save feature importance rankings graph
        plt.figure(figsize=(10, 6))
        sns.barplot(x='Importance', y='Feature', data=importance_df, hue='Feature', palette='viridis', legend=False)
        plt.title('Champion Random Forest Model - Feature Importance Rankings', fontsize=14, pad=15)
        plt.xlabel('Relative Gini Importance Score')
        plt.ylabel('Engineered Model Feature')
        plt.tight_layout()
        plt.savefig(os.path.join(self.output_dir, 'charts', 'v5_feature_importance.png'), dpi=300)
        plt.close()
        print("[*] Gini weight extraction complete. Saved asset as: 'v5_feature_importance.png'")

    def mock_production_endpoint(self, raw_unseen_payload):
        """
        Step 5: Production Gateway Deployment Emulation (LO5)
        Uses the correct step identifier ('prep') to resolve transformer properties.
        """
        input_df = pd.DataFrame([raw_unseen_payload])
        
        # Sync input data layer features with the exact transformer configuration step ('prep')
        required_cols = self.champion_pipeline.named_steps['prep'].transformers_[0][2]
        for col in required_cols:
            if col not in input_df.columns: 
                input_df[col] = 0.0
            
        predicted_class = self.champion_pipeline.predict(input_df)[0]
        confidence_score = self.champion_pipeline.predict_proba(input_df)[0][1]
        
        state_label = "CRITICAL: NETWORK DEMAND SURGE DETECTED" if predicted_class == 1 else "NORMAL: STABLE REVENUE RUNTIME"
        return {
            "Classification_Class": int(predicted_class),
            "Operational_Status": state_label,
            "Surge_Probability_Score": round(float(confidence_score), 4)
        }

# =====================================================================
# METRIC ANALYSIS CONTROL GATEWAY
# =====================================================================
if __name__ == "__main__":
    target_filename = "opal_trips.csv"
    
    if os.path.exists(target_filename):
        print(f"[+] Active data layer verified. Ingesting from local path: '{target_filename}'")
        pipeline = TfNSWTransitClassificationPipeline(target_filename)
        
        # Execute pipeline steps sequentially
        pipeline.extract_and_clean_data()
        pipeline.engineer_advanced_features()
        pipeline.execute_exploratory_visuals()
        pipeline.run_isolated_modeling()
        
        # --- SIMULATE LIVE DEPLOYMENT ---
        print("\n" + "="*70)
        print("[STAGE 5] TESTING INTERACTIVE SIMULATED API ENDPOINT GATEWAY")
        print("="*70)
        
        # Build dynamic real-time mock payload matched with historical features
        first_mode = pipeline.modal_cols[0] if pipeline.modal_cols else 'Bus'
        mock_live_payload = {
            first_mode: 25000000.0,
            'Total_Lag_1M': 75000000.0,
            'Total_Lag_2M': 71000000.0,
            'Rolling_Mean_3M': 73000000.0,
            'Quarter': 'Q1'
        }
        
        # Fill in any other secondary numeric features with default values
        for mode in pipeline.modal_cols:
            if mode not in mock_live_payload: 
                mock_live_payload[mode] = 1000000.0
            
        api_response = pipeline.mock_production_endpoint(mock_live_payload)
        print(f"[*] API Gateway payload received:\n{mock_live_payload}")
        print(f"[*] Production Endpoint Response Message Block:\n{api_response}")
        print("="*70)
    else:
        print(f"[!] Error: Unable to locate source data file '{target_filename}'. Verify path parameters.")
```
